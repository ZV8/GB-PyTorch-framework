#### Нашел значение W1 в функции для произвольной выборки данных с помощью pytorch и обратного распространения ошибки минимизируя среднеквадратическую ошибку (RMSE)

```python
import numpy as np
import matplotlib.pyplot as plt
from torch import nn
from torch.functional import F
from copy import copy
import seaborn as sns
sns.set_style("whitegrid")
n = 1000
noise = torch.Tensor(np.random.normal(0, 0.05, size=n))
x = torch.arange(0, 1, 1/n)
w1 = torch.FloatTensor([0.5])
y = x.pow(w1) - 2*x**2 + 5 + noise
plt.figure(figsize=(14, 7))
plt.scatter(x, y, alpha=0.4)
```

```python
class Model(nn.Module):
    """Custom Pytorch model for gradient optimization.
    """
    def __init__(self):
        
        super().__init__()
        # initialize weights with random numbers
        weights = torch.distributions.Uniform(0, 0.1).sample((1,))
        # make weights torch parameters
        self.weights = nn.Parameter(weights)        
        
    def forward(self, X):
        """Implement function to be optimised. In this case, function (x**w1 - 2 * x**2 + 5)
        """
        w1 = self.weights
        return X.pow(w1) - X.pow(2).mul(2) + 5
```

```python
def training_loop(model, optimizer, n=50):
    "Training loop for torch model."
    losses = []
    for i in range(n):
        preds = model(x)
        loss = F.mse_loss(preds, y).sqrt()
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        losses.append(loss)  
    return losses
```

#### Отобразим графики предсказанных значений функции с найденными весами в зависимости от количества итераций обратного распространения ошибки.

```python
fig = plt.figure(figsize=(20, 60))
total_axes = 20
subplt = 1

for n in range(10, 210, 10):
    m = Model()
    opt = torch.optim.Adam(m.parameters(), lr=0.01)
    losses = training_loop(m, opt, n=n)
    preds = m(x)

    ax = fig.add_subplot(total_axes, 2, subplt) # добавляем axes для каждого графика
    ax.scatter(x, preds.detach().numpy())
    ax.scatter(x, y, alpha=.3)
    ax.title.set_text(f'После {n} итераций оптимизации.')
    subplt+=1

    ax2 = fig.add_subplot(total_axes, 2, subplt) # добавляем axes для каждого графика
    ax2.plot(losses)
    ax2.title.set_text(f'W1 = {m.weights.detach().numpy()[0]}')
    subplt+=1

plt.tight_layout()
plt.show()
```

#### На графиках видим, как после 100 итераций значение веса стабилизируется.

```python
print(f'Итоговое значение W1: {m.weights.detach().numpy()}')
```

```python
import numpy as np
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

training_data = datasets.MNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor()
)

test_data = datasets.MNIST(
    root="data",
    train=False,
    download=True,
    transform=ToTensor()
)

train_dataloader = DataLoader(training_data, batch_size=64)
test_dataloader = DataLoader(test_data, batch_size=64)
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f'Using {device} device')
```

```python
model = nn.Sequential(
    nn.Flatten(),
    nn.Linear(in_features=28*28, out_features=128),
    nn.ReLU(),
    nn.Dropout2d(0.2),
    nn.Linear(128, 10)
).to(device)
print(model)
```

```python
def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    for batch, (X, y) in enumerate(dataloader):
        # Compute prediction and loss
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), batch * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")


def test_loop(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0

    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()

    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

```python
learning_rate = 1e-2
batch_size = 64
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

epochs = 5
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loop(train_dataloader, model, loss_fn, optimizer)
    test_loop(test_dataloader, model, loss_fn)
print("Done!")
```
